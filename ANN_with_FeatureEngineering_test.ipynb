```python
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(TrainData)
```

```python
rescaledX1 = tf.convert_to_tensor(
    rescaledX,
    dtype=None,
    dtype_hint=None,
    name=None
)
```

```python
TrainLabels1 = TrainLabels.astype('float64')
```

```python
TrainLabels2 = tf.convert_to_tensor(
    TrainLabels1,
    dtype=None,
    dtype_hint=None,
    name=None
)
```

```python
model = keras.Sequential([
#    keras.layers.Flatten(input_shape=(1, 6)),
#    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(22, activation=tf.nn.relu), 
    keras.layers.Dense(100, activation=tf.nn.relu),
    keras.layers.Dense(120, activation=tf.nn.relu),
    keras.layers.Dense(2, activation=tf.nn.softmax) #2=number of categorical values in categorical dependent variable
])
```

```python
model.compile(
              optimizer='adam',
#             optimizer='sgd', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(rescaledX1, TrainLabels2, epochs=90) 
```

```python
train_loss, train_acc = model.evaluate(rescaledX1, TrainLabels2)

print('Accuracy:', train_acc)
print('Loss:', train_loss)
```
